```python
import numpy as np
questions['conversion_asktake'] =  np.where(questions['TakenByAttorneyUno'].notnull(), 1, 0)

questions['conversion_askclose'] = np.where(questions['ClosedByAttorneyUno'].notnull(), 1, 0)

```

```python
questions
```

```python
state_data = questions.groupby('StateAbbr').agg(
    total_questions=('StateAbbr', 'count'),
    conversion_asktake=('conversion_asktake', 'sum'),
    conversion_askclose=('conversion_askclose', 'sum')
)


state_data['takerate'] = state_data['conversion_asktake'] / state_data['total_questions']
state_data['closerate'] = state_data['conversion_askclose'] / state_data['total_questions']

state_data
```

```python
totalquestions_rate = state_data
totalquestions_rate.to_csv("totalquestions_rate.csv")
```

```python
questions['conversion_notanswbutclosed'] = (questions['conversion_asktake'] == 0) & (questions['conversion_askclose'] == 1)
questions['conversion_notanswbutclosed'] = questions['conversion_notanswbutclosed'].astype(int)
```

```python
questions['YearMonth'] = questions['AskedOnUtc'].dt.strftime('%Y-%m')
```

```python
state_month_data = questions.groupby(['StateAbbr', 'YearMonth']).agg(
    total_questions=('QuestionUno', 'count'),
    conversion_asktake=('conversion_asktake', 'sum'),
    conversion_askclose=('conversion_askclose', 'sum'),
    total_notakeclosed=('conversion_notanswbutclosed', 'sum')
).reset_index()
```

```python
state_month_data['rec_ask'] = state_month_data['conversion_asktake']/state_month_data['total_questions']
```

```python
state_month_data_sorted = state_month_data.sort_values('YearMonth', ascending=True)
state_month_data_sorted
```

```python
state_month_data_sorted
```

```python
state_month_data_sorted['YearMonth'] = pd.to_datetime(state_month_data_sorted['YearMonth'])
state_month_data_sorted = state_month_data_sorted.sort_values('YearMonth', ascending=True)
state_month_data_sorted['YearMonth'] = state_month_data_sorted['YearMonth'].dt.strftime('%m-%Y')
state_month_data_sorted
```

```python
import pandas as pd 
import plotly.express as px


fig = px.choropleth(state_month_data_sorted,
                    locations='StateAbbr', 
                    locationmode="USA-states", 
                    color='rec_ask',
                    color_continuous_scale="Viridis_r", 
                    scope="usa",
                    animation_frame='YearMonth',
                    range_color=[min(state_month_data_sorted['rec_ask']), max(state_month_data_sorted['rec_ask'])])
fig.layout.updatemenus[0].buttons[0].args[1]['frame']['duration'] = 100
fig.update_layout(
    title='Monthly change of taken questions rate in different states between 2012 and 2022'
)
fig.show()
```
